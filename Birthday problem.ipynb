```python
import numpy as np
import matplotlib.pyplot as plt
```

#### Analytical solution

Expression for probability of no match
$$P_{nomatch} = \prod_{i=0}^m \frac{{period} - i}{period}$$
$$P_{match} = 1-P_{nomatch}$$



Case for maximum 23 agents

```python
period = 365
n_agents = 23
agents_sequence = np.arange(n_agents)
P_no_match = np.prod((period - agents_sequence) / period)
P_match = 1 - P_no_match
print(f"Probability of no matches {P_no_match}")
print(f"Probability of matches {P_match}", )
```

#### Probability for different number of agents

```python
period = 365
cases = list(range(1, 90, 2))
Ps = []
print(f"Probability of at least two matches in a room")

for n_agents in cases:
    agents_sequence = np.arange(n_agents)
    P_no_match = np.prod((period - agents_sequence) / period)
    P_match = 1 - P_no_match
    Ps.append(np.round(P_match, 4))
    print(f"If {n_agents} people: {np.round(P_match, 4)}")
```

##### Cumulative distribution function

```python
plt.figure(facecolor="silver", edgecolor="black")

_ = plt.plot(cases, Ps, label="Accumulated probability")
_ = plt.plot(23, Ps[cases.index(23)], "ro")
_ = plt.plot(41, Ps[cases.index(41)], "ro")

plt.annotate(text="0.507 if 23 people", xy=(23, 0.45), xycoords="data")
plt.annotate(text="0.903 if 41 people", xy=(40, 0.85), xycoords="data")

plt.xlabel("People in the room")
plt.ylabel("Probability")
plt.legend()
```
