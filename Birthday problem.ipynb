```python
cs = np.arange(1, 90, 2).tolist()
Ps = []
days_year = 365

print(f"Probability of at least two coincidences in a room")

for c in cs:
    x = np.arange(c)
    P_n_coincidence = np.prod((days_year - x) / days_year)
    P_y_coincidence = 1 - P_n_coincidence
    P_y_coincidence_print = np.round(P_y_coincidence, 4)
    Ps.append(P_y_coincidence_print)
    print(f"If {c} people: {P_y_coincidence_print}")
```

```python
plt.figure(facecolor="silver", edgecolor="black")

_ = plt.plot(cs, Ps, label="Accumulated probability")
_ = plt.plot(23, Ps[cs.index(23)], "ro")
_ = plt.plot(41, Ps[cs.index(41)], "ro")

plt.annotate(text="~0.51 for 23 people", xy=(23, 0.45), xycoords="data")
plt.annotate(text="0.9 for 41 people", xy=(40, 0.85), xycoords="data")

plt.xlabel("People in the room")
plt.ylabel("Probability")
plt.legend()
```
